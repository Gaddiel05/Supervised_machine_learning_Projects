```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('Data/sonar.all-data.csv')
```

```python
df.head()
```

```python
df['Label'].value_counts()
```

```python
sns.heatmap(df.drop('Label', axis=1).corr(), cmap="coolwarm");
```

```python
df['Target'] = df['Label'].map({'R':0, 'M':1})
```

```python
df.head()
```

```python
df.drop('Label', axis=1).corr()['Target']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import ConfusionMatrixDisplay, classification_report, r2_score, accuracy_score
```

```python
X = df.drop(['Label','Target'], axis=1)
```

```python
y = df['Label']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

```python
scaler = StandardScaler()
```

```python
knn = KNeighborsClassifier()
```

```python
knn.get_params().keys()
```

```python
operartions = [('scaler', scaler), ('knn',knn)]
```

```python
pipe = Pipeline(operartions)
```

```python
np.sqrt(len(df))
```

```python
knn_values = list(range(1,21))
```

```python
params = {'knn__n_neighbors': knn_values}
```

```python
grid_model = GridSearchCV(pipe, param_grid=params, scoring='accuracy', cv=5)
```

```python
grid_model.fit(X_train, y_train)
```

```python
grid_model.best_params_
```

```python
#grid_model.best_estimator_.get_params()
```

```python
MTS = grid_model.cv_results_['mean_test_score']
MTS
```

```python
sns.lineplot(MTS)
plt.xlabel('K')
plt.ylabel('Accuracy')
plt.xlim(-1,21)
```

### 2nd way of finding k (Elbo method)

```python
scaled_X_train = scaler.fit_transform(X_train)
```

```python
scaled_X_test = scaler.transform(X_test)
```

```python
test_error_rates = []


for k in range(1,30):
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(scaled_X_train,y_train) 
   
    y_pred_test = knn_model.predict(scaled_X_test)
    
    test_error = 1 - accuracy_score(y_test,y_pred_test)
    test_error_rates.append(test_error)
```

```python
plt.figure(figsize=(10,6),dpi=200)
plt.plot(range(1,30),test_error_rates,label='Test Error')
plt.legend()
plt.ylabel('Error Rate')
plt.xlabel("K Value")
```

```python
final_model = KNeighborsClassifier(n_neighbors=1)
```

```python
operations = [('scaler',scaler), ('knn',final_model)]
```

```python
pipe = Pipeline(operations)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
y_pred
```

```python
print(classification_report(y_test, y_pred))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred);
```
